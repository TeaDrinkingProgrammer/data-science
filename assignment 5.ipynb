# Filtering and univariate analysis on lifexpectancy dataset

```python
import pandas as pd
```

```python
lifeExpectancy = pd.read_csv('lifeExpectancyAtBirth.csv', sep=',')
```

```python
lifeExpectancy.head()
lifeExpectancy
```

```python
filter = (lifeExpectancy['Dim1'] == 'Both sexes') & (lifeExpectancy['Period'] == 2019)
filteredDf = lifeExpectancy[filter]
filteredDf
```

```python
filteredDf['First Tooltip'].plot(kind='box')
```

## The countries with the highest and lowest life expectancy

```python
filteredDf.sort_values(by='First Tooltip',ascending= False)
```

```python
print(filteredDf.describe())
```
